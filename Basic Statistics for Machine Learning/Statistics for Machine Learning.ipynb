**Mean:** Mean is the arithmetic average of a data set.

**Median:** The median is the middle number in a data set when the numbers are listed in either ascending or descending order.

**Mode:** The mode is the value that occurs the most often in a data set.

**Population:** A population is the entire group that you want to draw conclusions about.

**Sample:** A sample is the specific group that you will collect data from. The size of the sample is always less than the total size of the population.

**Mean:**
- Population (μ) = $\frac{\sum \limits_{N}^{i=1} x_{i}}{N}$

- Sample ($\hat{X}$) = $\frac{\sum \limits_{n}^{i=1} x_{i}}{n}$

**Standard Deviation:**
- Population (σ) = $\sqrt{\frac{\sum \limits_{N}^{i=1} (x_{i}-μ)^2}{N}}$

- Sample (s) = $\sqrt{\frac{\sum \limits_{n}^{i=1} (x_{i}-\hat{x})^2}{n-1}}$

where<br>
&emsp; N = number of items in population<br>
&emsp; n = number of items in sample<br>

### Normal Distribution:
Normal distribution, also known as the Gaussian distribution, is a probability distribution that is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean. In graph form, normal distribution will appear as a bell curve.

For a Gaussian or Normal Distribution, μ need not be 0 and σ need not be 1.

For a Standard Normal Distribution, μ=0 and σ=1.

**Empirical Rule:**

Pr(μ-σ ≤ X ≤ μ+σ) ≈ 68.27%

Pr(μ-2σ ≤ X ≤ μ+2σ) ≈ 95.45%

Pr(μ-3σ ≤ X ≤ μ+3σ) ≈ 99.73%

### Log Normal Distribution:

If the random variable X is log-normally distributed, then Y = ln(X) has a normal distribution.

### Z Score:
Z Score values tells you how many standard deviations you are away from the mean. It is used to convert Normal Distribution into Standard Normal Distribution.
$$ Z=\frac{X-μ}{σ}$$

A Z-Score table, is a table that allows us to know the percentage of values below (to the left) a z-score in a standard normal distribution.

### Probability Density Function:
The Probability Density Function(PDF) is the probability function which is represented for the density of a continuous random variable lying between a certain range of values.

In case of a continuous random variable, the probability taken by X on some given value x is always 0. In this case, if we find P(X = x), it does not work. Instead of this, we require to calculate the probability of X lying in an interval (a, b). Now, we have to calculate it for P(a< X< b). This can be done by using a PDF. The Probability distribution function formula is defined as,
$$P(a<X<b)=\int_{a}^{b} f(x)$$

### Cumulative Distribution Function:
The Cumulative Distribution Function (CDF) of a random variable X is denoted by F(x), and is defined as F(x) = P(X ≤ x). It is used to describe the probability distribution of random variables in a table.

The CDF defined for a continuous random variable is given as,
$$F(x)=P(X ≤ x)=\int_{-\infty}^{x} f(x)dx$$
$$F(b)−F(a)=P(a<X≤b)=\int_{a}^{b} f(x)$$

##### Different measures of correlation:
1. Covariance
2. Pearson Correlation Coefficient
3. Spearman's Rank Correlation Coefficient

### 1. Covariance:
Covariance is a measure of the joint variability of two random variables.

$$Cov(X,Y)=\frac{\sum \limits_{n}^{i=1} (x_{i}-\hat{x})(y_{i}-\hat{y})}{n}$$


<img src="attachments/positive-negative-corrrelation-examples.jpg" width="500"/>

### 2. Pearson Correlation Coefficient:
The Pearson Correlation Coefficient ranges from −1 to 1.

$$ρ_{X,Y}=\frac{Cov(X,Y)}{σ_{X}σ_{Y}}$$


<img src="attachments/pearson-correlation-coefficient-values.png" width="400"/>

### 3. Spearman's Rank Correlation Coefficient:
The Spearman Correlation between two variables is equal to the Pearson correlation between the rank values of those two variables.

$$r_{s}=ρ_{r_{X},r_{Y}}=\frac{Cov(r_{X},r_{Y})}{σ_{r_{X}}σ_{r_{Y}}}$$
where<br>
&emsp; $r_{s}$ denotes Spearman's Rank Correlation Coefficient<br>
&emsp; $r_{X},r_{Y}$ denotes rank variables of X and Y<br>
&emsp; $ρ_{r_{X},r_{Y}}$ denotes Pearson Correlation Coefficient of rank variables<br>
&emsp; $Cov(r_{X},r_{Y})$ denotes Covariance of rank variables<br>
&emsp; $σ_{r_{X}},σ_{r_{Y}}$ denotes standard deviations of rank variables<br>

##### What is an Outlier?
An Outlier is a data point in a dataset that is distant from all other observations. A data point that lies outside the overall distribution of the dataset.

##### What are the criteria to identify an Outlier?
1. Data point that falls outside of 1.5 times of an interquartile range above the 3rd quartile and below the 1st quartile.
2. Data point that falls outside of 3 standard deviations. Implement Z Score on all data points and if Z Score of any data point is outside of 3, it will be considered as an outlier.

##### What are the impactsof having outliers in a dataset?
1. It causes various problems during statistical analysis.
2. It may cause significant impact on the mean and standard deviation.

**Interquartile Range:** Interquartile Range abbreviated as IQR, is the range from the 25th percentile to the 75th percentile of any given data set.
#### Steps to find interquartile range:
1. Sort given data in ascending order, find the median which is Q2.
2. Split the sorted data into two clusters based on median and find median for each cluster which is Q1 and Q3.
3. IQR = Q3-Q1.

where Q1 is the “middle” value in the first half of the rank-ordered data set, Q2 is the median value in the set, Q3 is the “middle” value in the second half of the rank-ordered data set.

#### Steps to find outliers using interquartile range:
1. Lower Bound = Q1-1.5*IQR
2. Upper Bound = Q3+1.5*IQR
3. Data points which don't fall in between lower and upper bound are considered as outliers.
