# 1. MatplotLib:
Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy.<br>
plt.show() is not required when "%matplotlib inline" is executed, else plt.show() should be executed for seeing plot.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

#### a. Scatter Plot:
A graph in which the values of two variables are plotted along two axes, the pattern of the resulting points revealing any correlation present. It consists of points.

```python
# initializing variables to plot
x=np.arange(0,10)
y=np.arange(11,21)
# Plot x and y on graph, c='g' is the colour of plot
plt.scatter(x,y,c='g')
# Labelling X Axis
plt.xlabel('X Axis')
# Labelling Y Axis
plt.ylabel('Y Axis')
# Adding Title to the plot
plt.title('plt.scatter()')
```

#### b. Plot:
A graph in which the values of two variables are plotted along two axes. It consists of line.

```python
# initializing variables to plot
x=np.arange(0,10)
y=x*x
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('plt.plot()')
```

#### c. Subplots:
Create multiple graphs or subplots in same canvas.

```python
# nrows=2, ncols=2 and index=1
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
# nrows=2, ncols=2 and index=2
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
# nrows=2, ncols=2 and index=3
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
# nrows=2, ncols=2 and index=4
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
# Pi value
np.pi
```

```python
# x ranges from 0 to 4, with increment of 0.1
x = np.arange(0, 4 * np.pi, 0.1)
# fetching sin(x) value for all x values
y = np.sin(x)
# plotting x and y
plt.plot(x, y)
plt.title("Sine Wave Form")
```

#### d. Bar Plot:
A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent.

```python
x = ["a","b","c"]  
heights = [2,8,10] 
plt.bar(x,heights)
plt.title('Bar Graph')
plt.xlabel('X Axis')
plt.ylabel('Height Axis')
```

#### e. Histogram:
A diagram consisting of rectangles whose area is proportional to the frequency of a variable and whose width is equal to the class interval.

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a,bins=20)
plt.title("histogram")
```

#### f. Box Plot:
A box plot is a method used for explanatory data analysis to visually show the distribution of numerical data and skewness through their data quartiles or percentiles.

<img src="attachments/box plot.png" width="450">

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=False)
plt.show()
```

# 2. Seaborn:

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df.head()
```

#### Correlation:
Compute pairwise correlation of integer or float columns, excluding NA/null values.

```python
df.corr()
```

#### a. Correlation with Heatmap (Numeric Plot):
A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix between two columns. It is very important in Feature Selection.

```python
# annot=True for displaying correlation value on figure
sns.heatmap(df.corr(),annot=True)
```

#### b. Joint Plot (Numeric Plot):
A joint plot allows to study the relationship between two numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot. The Graphs on top and right are histograms of the numeric variables.

```python
# kind = "hex", hexagonal bins
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
# kind = regression, draws a probability density function for histograms and
# a regression line for scatter plot
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

#### c. Pair plot (Numeric Plot):
A pair plot is also known as a scatterplot, in which every numeric variable is mapped to every other numeric variable, to form a grid plot of all combinations.

```python
sns.pairplot(df)
```

```python
# Hue can be any categorical variable to distinguishing during analysis 
sns.pairplot(df,hue='sex')
```

#### d. Dist Plot (Numeric Plot):
Dist plot is histogram.

```python
sns.distplot(df['tip'])
```

```python
# kde means kernel density estimator
sns.distplot(df['tip'],kde=False,bins=10)
```

#### e. Count Plot (Categorical Plot):
Show the counts of observations for each categorical value using bars.

```python
sns.countplot('sex',data=df)
```

#### f. Bar Plot (Categorical Plot):
Group records by categories in x, impose estimator on y for each group and plot.<br>
Default Estimator is Mean.

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
df.groupby("sex")["total_bill"].mean()
```

#### g. Box Plot (Categorical Plot):
Group records based on a categorical variable, and plot distribution of a numerical variable for each group.

```python
sns.boxplot('smoker','total_bill', data=df)
```

```python
sns.boxplot(x="total_bill", y="day", data=df,hue="sex")
```

#### h. Violin Plot:
Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot. It is a combination of Boxplot and Kernal Density Estimator.

```python
sns.violinplot(x="total_bill", y="day", data=df)
```

# Types of Data Analysis:
- Univarient Analysis
- Bivarient Analysis
- Multivarient Analysis

### i. Univarient Analysis:
Univariate analysis is the technique of analyzing data. Uni means one, so in other words your data has only one variable. It takes data, summarizes that data and finds patterns in the data.<br>
Eg: Countplot, Barplot, Distplot (Histogram)

```python
df=sns.load_dataset("iris")
df.head()
```

```python
# getting records with species as setosa
df_setosa=df.loc[df['species']=='setosa']
```

```python
df_virginica=df.loc[df['species']=='virginica']
df_versicolor=df.loc[df['species']=='versicolor']
```

```python
# plotting sepal_length feature of each species with marker 'o'
# np.zeros_like returns an array of zeros with the same shape and type as a given array
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'o')
plt.xlabel('Petal length')
plt.show()
```

### ii. Bivariate Analysis:
Bivariate analysis is where two variables are observed. One variable here is dependent while the other is independent. These variables are usually denoted by X and Y. So, here we analyse the changes occured between the two variables and to what extent.<br>
Eg: Jointplot, Scatterplot, Boxplot

```python
df=sns.load_dataset("iris")
df.head()
```

```python
sns.scatterplot(df["petal_length"],df["sepal_width"],hue=df["species"])
```

### iii. Multivariate Analysis:
Multivariate analysis is a method that examine patterns in multidimensional data by considering, at once, several data variables. It is an expansion of bivariate data analysis, which considers only two variables in its models. 

```python
df=sns.load_dataset("iris")
df.head()
```

```python
sns.pairplot(df,hue="species")
```
